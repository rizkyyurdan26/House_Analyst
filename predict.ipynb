```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine('mysql+pymysql://root:@localhost:3306/housedb')
query = 'select * from house'
df = pd.read_sql(query, con=engine)

pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_colwidth', None)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd

X = df[['buildingArea', 'landArea', 'bedroom', 'bathroom', 'garage']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

model = LinearRegression()

# Train model
model.fit(X_train, y_train)

#  R² score
r2_score = model.score(X_test, y_test)
print(f"R² Score: {r2_score}")

predictions = model.predict(X_test)

predicted_vs_actual = X_test.copy() 
predicted_vs_actual['Predicted'] = predictions  
predicted_vs_actual['Actual'] = y_test 

print('\nPrice Predicted')
predicted_vs_actual.head()
```

```python
# === new data===
new_data = pd.DataFrame([{
    'buildingArea': 120,
    'landArea': 150,
    'bedroom': 3,
    'bathroom': 2,
    'garage': 1
}])

predicted_price = model.predict(new_data)[0]
print(f"\nPredicted Price house: Rp {predicted_price:,.2f}")
```
